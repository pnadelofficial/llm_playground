Below is an environment designed to interact with open source large language models on the HPC. 

### Software  

This example is using the model [`qwen15-72B-chat`](https://huggingface.co/Qwen/Qwen1.5-72B-Chat), which is a very powerful model based that reaches performance surpassing GPT3.5 and rivaling GPT4. ### To get started
1. Go to `Cell` in the ribbon above (between `Insert` and `Kernel`)
2. Click on `Run All` 
3. Wait for the checkpoint shards to be loaded (the progress bar below)
4. Start chatting by entering a message in the text area and clicking "Submit message"
5. To reset your chat, click on the refresh button just below `Kernel` and follow steps 1 - 4
